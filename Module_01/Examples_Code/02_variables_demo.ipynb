# Purpose: Understanding Variables and Data Types

```python
name = "Vishnu" # String (str)
age = 25 # integer (int)
fee = 5000.50 # Float (float)
is_Student = True # Boolean (bool)

print("Name:",name, "| Type:",type(name))
print("Age:",age, "| Type:",type(age))
print("fee:",fee, "| Type:",type(fee))
print("Is Student:",is_Student, "| Type:",type(is_Student))
```
