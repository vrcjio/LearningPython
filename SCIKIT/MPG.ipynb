```python
import pandas as pd
import seaborn as sns
from sklearn.impute import KNNImputer

# =====================================================================
# STEP 1: Data Load Karna Aur Missing Rows Ka Index Nikalna
# =====================================================================
df = sns.load_dataset('mpg')

print("="*70)
print("📌 STEP 1: DATA CLEANING SE PEHLE (ORIGINAL MISSING VALUES)")
print("="*70)
null_indices = df[df['horsepower'].isnull()].index
print(df.loc[null_indices, ['name', 'cylinders', 'displacement', 'weight', 'horsepower']])


# =====================================================================
# STEP 2: SPECIFICATION KE BASIS PAR KNN IMPUTER CHALANA
# =====================================================================
# Hum sirf wahi numeric columns select kar rahe hain jo car ki specification batate hain
spec_columns = ['cylinders', 'displacement', 'weight', 'acceleration', 'horsepower']

# KNNImputer initialize karna. 
# n_neighbors=3 ka matlab hai yeh 3 sabse milti-julti specifications wali cars ka average nikalega
imputer = KNNImputer(n_neighbors=3)

# Imputer ko specifications wale data par chalana (Yeh matrix return karta hai)
imputed_data = imputer.fit_transform(df[spec_columns])

# Matrix ko wapas ek temporary DataFrame me convert karna
df_imputed_specs = pd.DataFrame(imputed_data, columns=spec_columns)

# Original dataframe ke horsepower column ko is super-accurate specification-based horsepower se replace kar dena
df['horsepower'] = df_imputed_specs['horsepower']

# =====================================================================
# STEP 3: FINAL CHECK - Kya Specifications Ke Hisab Se Data Sahi Bhara?
# =====================================================================
print("\n" + "="*70)
print("📌 STEP 3: DATA CLEANING KE BAAD (KNN SPECIFICATION-BASED VALUES)")
print("="*70)
# Wahi 6 rows dobara check karte hain ki unme ab kya values aayi hain
print(df.loc[null_indices, ['name', 'cylinders', 'displacement', 'weight', 'horsepower']])

print(f"\nTotal Null Values left in Horsepower: {df['horsepower'].isnull().sum()}")
```
