```python
from sklearn.datasets import fetch_california_housing

housing_raw = fetch_california_housing()
print(housing_raw.keys())
```

```python
import pandas as pd

df = pd.DataFrame(data=housing_raw.data,columns=housing_raw.feature_names)
df['MaidHouseValue'] = housing_raw.target
print(df.info())
print(df)
# MedInc: Median Income (Logon ki kamai - Sabse important feature)
# HouseAge: Ghar kitne saal purane hain
# AveRooms / AveBedrms: Average kitne kamre hain
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

X = df[housing_raw.feature_names].to_numpy()
y = df['MaidHouseValue'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42,test_size=0.2)

print("\nModel is train... please wait 5-10 seconds...")
print("training...")
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\n" + "=" * 60)
# MAE batata hai ki dollar me hamara prediction average kitna upar-neeche hai
mae = mean_absolute_error(y_test, y_pred)
# R2 Score batata hai ki model kitna achha hai (Maximum 1.0 ho sakta hai)
r2 = r2_score(y_test, y_pred)
#model = RandomForestRegressor(n_estimators=50, random_state=42, n_jobs=-1)
print("="*60)
print(
    f"Mean Absolute Error (MAE): ${mae*100000:.2f}"
)# 100k se multiply kiya real dollar ke liye

print(f"🎯 R2 Score (Accuracy standard): {r2 * 100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, min_samples_split=5, n_jobs=-1, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print(
    f"Mean Absolute Error(MAE): {mae*100000:.2f}"
)
r2 = r2_score(y_test, y_pred)
print(f"🎯 R2 Score (Accuracy standard): {r2 * 100:.2f}%")
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

model = HistGradientBoostingRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(
    f"Mean Absolute Error(MAE): {mae*100000:.2f}"
)
r2 = r2_score(y_test, y_pred)
print(f"🎯 R2 Score (Accuracy standard): {r2 * 100:.2f}%")
```
