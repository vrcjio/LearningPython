## 1. Machine Learning Kya Hai? (What is Machine Learning?)

**Machine Learning (ML)** computer science aur Artificial Intelligence (AI) ki ek sub-field hai. Iska sabse bada maqsad yeh hai ki computers ko bina kisi explicit programming (har cheez ke liye alag se code likhe bina) ke data se seekhna (learn karna) sikhaya ja sake.

### Traditional Programming vs Machine Learning
* **Traditional Programming (Purana Tarika):** Aap computer ko **Data** dete ho + **Rules (Code)** dete ho = Computer aapko **Output** deta hai.
* **Machine Learning (Naya Tarika):** Aap computer ko bohot saara **Data** dete ho + uske sath **Outputs (Results)** bhi dete ho = Computer khud ba khud un dono ke beech ka **Rules (Pattern/Formula)** dhoodh leta hai.

## 2. Supervised vs Unsupervised Learning

Machine Learning ko algorithms ke seekhne ke tarike ke basis par alag-alag types mein baanta gaya hai. Isme do sabse bade aur important types hain:

### A. Supervised Learning (Dekh-rekh mein seekhna)
Is type ki learning mein hum model ko **Labeled Data** dete hain. Labeled data ka matlab hai ki computer ko input ke sath-sath uska sahi answer (output label) bhi pehle se bataya jata hai.
* **Kaise kaam karta hai:** Jaise ek teacher bachhe ko pehle question aur answer dono sikhata hai, aur fir exam mein naya question poochta hai.
* **Sub-types:** Iske do main types hote hain:
  1. **Classification:** Jab output koi category ya group ho (Jaise: Haan/Naa, Spam/Not Spam, Male/Female).
  2. **Regression:** Jab output koi continuous number ya value ho (Jaise: Price, Temperature, Age).

### B. Unsupervised Learning (Bina dekh-rekh ke seekhna)
Is type ki learning mein hum model ko **Unlabeled Data** dete hain. Isme computer ko sirf input diya jata hai, koi sahi answer ya label nahi diya jata. Computer ko khud hi data ke andar ka pattern dhoodhna padta hai.
* **Kaise kaam karta hai:** Jaise aapko kisi anjaan desh ke bohot saare kapde de diye jayein, aur aap unke rang ya kapde ke type ke hissab se unhe alag-alag dher (groups) mein baant dein, bina yeh jaane ki unhe kya kehte hain.
* **Sub-types:** Iske main types hote hain:
  1. **Clustering:** Ek jaise dikhne wale data ka group banana.
  2. **Dimensionality Reduction:** Data ke fuzool columns ko kam karna taaki calculation aasan ho sake.

### Quick Comparison Table:

| Feature | Supervised Learning | Unsupervised Learning |

## 3. Dataset Kya Hota Hai? (Understanding Datasets)

Machine Learning mein data hi sab kuch hai. **Dataset** ka matlab hota hai data ka ek collection (collection of data), jo aamtaur par ek table (rows aur columns) ke roop mein hota hai (jaise Excel sheet ya CSV file).

Ek ML dataset ko samajhne ke liye hume in terms ko samajhna hoga:

1. **Rows (Instances / Samples):** Har ek row ek single record ko darshati hai. Jaise agar dataset gharon ke baare mein hai, toh har ek row ek alag ghar ki jankari hogi.
2. **Columns (Features & Target):** Columns do tarah ke hote hain:
   * **Features (Independent Variables - $X$):** Yeh woh columns hote hain jo input ka kaam karte hain. Inhi ke aadhar par prediction ki jaati hai. (Example: Ghar ka size, floor number, locality, rooms ki sankhya).
   * **Target / Label (Dependent Variable - $y$):** Yeh woh column hota hai jise humara machine learning model predict karna chahta hai. (Example: Ghar ki Final Price).

### Dataset Split (Data ko Baantna):
Jab hume model train karna hota hai, toh hum poore dataset ko do hisson mein baant te hain:
* **Training Set (Aamtaur par 70-80%):** Is data ka use karke model ko sikhaya aur train kiya jata hai.
* **Testing Set (Aamtaur par 20-30%):** Is data ko model se chhupa kar rakha jata hai. Jab model seekh jata hai, toh is testing data par uska test liya jata hai ki woh kitna accurate perform kar raha hai.

## 5. Scikit-Learn Ke Core Features (4 Bade Kaam)

Scikit-learn ke andar pehle se hi hazaron lines ke mathematical codes ko functions ke roop mein pack kiya gaya hai. Yeh mainly 4 bade problems ko solve karta hai:

### A. Classification (Vargikaran)
* **Kya hai?** Data ko pehle se tay ki gayi categories ya classes mein divide karna.
* **Kyun use karein?** Jab aapko "Haan" ya "Naa", "A", "B" ya "C" jise discrete outputs chahiye hon.
* **Examples:**
  * Ek Bank check karna chahta hai ki loan lene wala aadmi default karega ya nahi.
  * Medical report dekh kar batana ki tumor malignant (cancerous) hai ya benign (normal).
* **Popular Algorithms:** * Logistic Regression (Naam mein regression hai par kaam classification ka karta hai)
  * Support Vector Machines (SVM)
  * Random Forest Classifier
  * Decision Trees

### B. Regression (Pratigaman)
* **Kya hai?** Continuous numerical values ko predict karna (yaani numbers ko predict karna jo badalte rehte hain).
* **Kyun use karein?** Jab output koi fix category nahi balki ek number hota hai.
* **Examples:**
  * Company ke purane records dekh kar agle mahine ki *Sales* predict karna.
  * Gold price ya stock prices ka prediction karna.
* **Popular Algorithms:**
  * Linear Regression
  * Ridge and Lasso Regression
  * Random Forest Regressor

### C. Clustering (Samoohikaran)
* **Kya hai?** Bina labels ke data points ko unki similarities (ek jaisi khoobiyon) ke aadhar par groups mein baantna.
* **Kyun use karein?** Jab aapko data ka structure samajhna ho par aapke paas koi answer key (label) na ho.
* **Examples:**
  * **Customer Segmentation:** Ek e-commerce company apne customers ko divide karti hai: "Kam kharch karne wale", "Bohot zyada kharch karne wale", aur "Sirf offers par khareedne wale". Isse marketing easy ho jati hai.
* **Popular Algorithms:**
  * K-Means Clustering
  * DBSCAN
  * Hierarchical Clustering

### D. Dimensionality Reduction (Dimensions Kam Karna)
* **Kya hai?** Kai baar datasets mein saikdon ya hazaron columns (features) hote hain. Bohot zyada columns hone se model bohot slow ho jata hai aur confuse ho jata hai (jise *Curse of Dimensionality* kehte hain). Yeh process fuzool ya milti-julti columns ko combine karke columns ki sankhya kam karta hai, bina important jankari khoye.
* **Examples:**
  * High-resolution images ke pixels ko kam karna taaki image recognition model fast chal sake.
* **Popular Algorithms:**
  * Principal Component Analysis (PCA)
  * t-SNE

## 6. Scikit-Learn Kyun Itna Popular Hai?

1. **Consistency & Simplicity:** Isme har algorithm ko chalane ka tarika bilkul same hai. Agar aapne ek algorithm seekh liya, toh aap baki ke saare algorithms bina naya code seekhe chala sakte hain.
2. **Data Preprocessing Tools:** ML model mein seedhe raw data nahi dala jata. Data ko saaf karna padta hai (missing values bharna, text ko numbers mein badalna). Sklearn yeh sab bohot aasani se kar deta hai (`SimpleImputer`, `OneHotEncoder`, `StandardScaler`).
3. **Model Evaluation:** Model kitna achha kaam kar raha hai yeh check karne ke liye isme kamaal ke tools hain, jaise Train-Test Split, Cross-Validation, Accuracy Score, Confusion Matrix.
4. **Open Source & Free:** Yeh bilkul free hai aur iski community bohot badi hai, jisse koi bhi error aane par solution internet par aasani se mil jata hai.
