```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# --- Step 1: Model ko Train karna (Pehle ki tarah) ---
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

print("🎉 Model Train ho chuka hai! Ab aap live test kar sakte hain.\n")
print("=" * 60)
print("🌸 NAYE PHOOL KI DETAILS ENTER KAREIN 🌸")
print("=" * 60)

# --- Step 2: User se Live Input lena ---
# Hint ke liye Iris dataset ki average range: 1.0 se 8.0 ke beech
try:
    sepal_l = float(input("1. Sepal Length (cm) daliye [e.g., 5.1]: "))
    sepal_w = float(input("2. Sepal Width (cm) daliye  [e.g., 3.5]: "))
    petal_l = float(input("3. Petal Length (cm) daliye [e.g., 1.4]: "))
    petal_w = float(input("4. Petal Width (cm) daliye  [e.g., 0.2]: "))

    # --- Step 3: Input ko Model ke format me badalna ---
    # Model hamesha 2D array (matrix) mangta hai, isliye humne ise [[]] me dala
    naya_phool = np.array([[sepal_l, sepal_w, petal_l, petal_w]])

    # --- Step 4: Prediction aur Output dikhana ---
    # Model hume 0, 1 ya 2 return karega
    prediction_code = model.predict(naya_phool)[0]

    # Code se phool ka asli naam nikalna
    phool_ka_naam = iris.target_names[prediction_code]

    # Model kitna confident hai (Probability percentage)
    probabilities = model.predict_proba(naya_phool)[0]
    confidence = probabilities[prediction_code] * 100

    print("\n" + "=" * 60)
    print(f"🤖 COMPUTER KA JAWAB: Ye phool 👉 **{phool_ka_naam.upper()}** hai!")
    print(f"🎯 Confidence Level: {confidence:.2f}%")
    print("=" * 60)

except ValueError:
    print("\n❌ Error: Kripya sirf numbers (points me) enter karein!")
```
